# 1.3 Meta Data Cleaning

```python
url_drop = [c for c in tracks.columns if 'url' in c] #All url information will not be useful for prediction
copyright_drop = [c for c in tracks.columns if 'copyright' in c] #all copyright information will not be useful for prediction
license_drop = [c for c in tracks.columns if 'license' in c] #All licensing information will not be useful for prediction
missing_cols = [c for c in tracks.columns if sum(~tracks[c].isnull()) < 100000] #columns with a lot of missing data
addl_not_useful = ['track_title', 'track_number', 'track_image_file', 
                   'track_disc_number', 'track_date_created', 'artist_id', 
                   'album_title', 'artist_name', 'track_file', 'tags', 
                   'album_id'] #additional columns that will not be useful for prediction

col_drop = url_drop + copyright_drop + license_drop + missing_cols + addl_not_useful
tracks.drop(columns=col_drop, inplace=True)
tracks.info()
```

Now things are starting to look a bit better, but we still need to do some work to convert the genre information and track duration to a number. As a note, the track id will also need to be deleted, but it will need to be used to connect our audio files to the meta features before it is removed.

```python
#Extracting the genre id from the genre column
#This code was taken and modified from https://github.com/mdeff/fma/blob/master/creation.ipynb
def convert_genres(genres):
    genres = ast.literal_eval(genres)
    return [int(genre['genre_id']) for genre in genres]

tracks['track_genres'].fillna('[]', inplace=True)
tracks['track_genres'] = tracks['track_genres'].map(convert_genres)
```

Now we have a list of strings of the genre IDs, but we have to figure out what to do with the tracks that have more than one genre. Do we create new columns so all of them can be turned into numbers, do we drop those rows, do we select one genre to use and if so how do we know it is a good selection? Let's first look at how many rows we have with more than one genre.

```python
num_genres = tracks['track_genres'].map(len)
print('number of tracks with more than one genre classification: {}'
      .format(sum(num_genres > 1)))

print('number of genres distribution')
print(num_genres.value_counts())
```

So most of the songs have more than one classification, with nearly half having 3 or more genres. Let's just keep the first three genres as seperate columns and drop any remaining genres.

```python
def first_genre(genre_list):
    if len(genre_list) > 0:
        genre_1 = genre_list[0]
    else:
        genre_1 = np.nan
    return genre_1

def second_genre(genre_list):
    if len(genre_list) > 1:
        genre_2 = genre_list[1]
    else:
        genre_2 = np.nan
    return genre_2

def third_genre(genre_list):
    if len(genre_list) > 2:
        genre_3 = genre_list[2]
    else:
        genre_3 = np.nan
    return genre_3

tracks['genre1'] = tracks['track_genres'].map(first_genre)
tracks['genre2'] = tracks['track_genres'].map(second_genre)
tracks['genre3'] = tracks['track_genres'].map(third_genre)

tracks.drop(columns=['track_genres'], inplace=True)
```

Lastly, we need to convert the song duration from a string of minute and seconds to an integer of seconds.

```python
def convert_times(times):
    if len(times) > 2:
        return 60*60*int(times[0]) + 60*int(times[1]) + int(times[2])
    else:
        return 60*int(times[0]) +  int(times[1])

tracks['track_duration'] = tracks['track_duration'].str.split(':').map(convert_times)
```

# 1.4 Meta Data EDA

Let's first look at the bitrate since it could potentially impact the songs popularity. A songs bitrate is related to the songs audio quality, where a higher bitrate means the song has a higher audio quality. Specifically, below 90 Kbps, humans can notice a significant drop in audio quality, which could impact their impression of the music. We can take a quick look to determine if this is true by ploting the number of streams vs. the bitrate.

```python
print('Number of songs with a bitrate < 90 Kbps: {}\n'.format(sum(tracks['track_bit_rate'] < 90000)))

fig = plt.figure(figsize=(10,8))
plt.hist(tracks['track_bit_rate'], bins=20)
plt.xlabel('Bitrate (bps)')
plt.ylabel('Count of Songs')
plt.title('Histogram of Sample Songs Bitrate')

plt.show()
```

```python
from sklearn.linear_model import LinearRegression

fit_data = tracks[['track_bit_rate', 'track_listens']].dropna()
lr = LinearRegression().fit(X=fit_data['track_bit_rate'].to_numpy().reshape(-1,1), y=fit_data['track_listens'])

print('correlation between bitrate and listens: {}'.format(lr.coef_[0]))

lr_bitrates = np.linspace(0,400000,20).reshape(-1,1)
lr_listens = lr.predict(lr_bitrates)

fig = plt.figure(figsize=(10,8))
plt.plot(tracks['track_bit_rate'], tracks['track_listens'], marker='o', linestyle=' ')
plt.plot(lr_bitrates, lr_listens)
plt.xlabel('Bitrate (bps)')
plt.ylabel('Song listens')

plt.show()
```

It looks like most songs have common bit rates around 320 Kbps, 256 Kbps, and 192 kKbps and there is no strong correlation between the bit rate and the number of times the song was played. However, this correlation is partially misleading because most songs don't become hits. Instead we should look at only the songs that are deemed hits, which we have yet to define. Should we use a threshold value, or a percentile?  
<span style="font-size:1.25em">__If we choose the 98 percentile of number of listens value, then our threshold value is 15,000 and there are only 2195 songs that meet that criteria.__</span>  
15,000 listens seems fairly low, but again the songs are from a free music repository so they are less likely to be played in general.

```python
top_hits = tracks[tracks['track_listens'] > np.percentile(tracks['track_listens'], [98])[0]]
top_hits = top_hits[['track_listens', 'track_bit_rate']].dropna()

lr = LinearRegression().fit(X=top_hits['track_bit_rate'].to_numpy().reshape(-1,1), y=top_hits['track_listens'])

print('correlation between bitrate and listens: {}'.format(lr.coef_[0]))

lr_bitrates = np.linspace(0,400000,20).reshape(-1,1)
lr_listens = lr.predict(lr_bitrates)

fig = plt.figure(figsize=(10,8))
plt.plot(top_hits['track_bit_rate'], top_hits['track_listens'], marker='o', linestyle=' ')
plt.plot(lr_bitrates, lr_listens)
plt.xlabel('Bitrate (bps)')
plt.ylabel('Song listens')

plt.show()
```

Again there is no correlation, or rather the correlation is small an slightly negative so we can rule out our hyptohesis that audio quality could impact the number of times a song was listend to. Alternatively, instead of using the raw audio we could create features from transformations and calculating primitives. Let's come up with a list of potential features we could calculate on the data.

Librosa has a feature library that calculates the following attributes of audio files:  
* Chromagram - related to the twelve different pitch classes, capture harmonic and melodic characteristics, specifically the energy of the pitches, returns an array  
* Constant-Q chromagram - signal processing method to transform a data series to the frequency domain, related to Fourier transform  
* Chroma Energy Normalized - takes statistical values over large windows, best used for audio matching and similarity  
* Mel-scaled spectogram - mel scale is a linear transform of the frequency, mel spectogram is a spectrogram with the mel scale on the y axis instead of the frequency  
* Mel-frequency cepstral coefficients - the coefficients that make up a representation of the short-term power spectrum. Often used in speech recognition, genre classification, and audio similarity  
* Root mean square - Computed on each frame  
* spectral centroid - idicates where the center of mass of the spectrum is sglocated, weighted mean of frequencies present in the signal.  
* spectral bandwidth - portion of the signal spectrum in the frequency domain that cotains most of the energy of signal
* spectral contrast - the difference between the peak and valley of a spectrum
* spectral flatness - also known as tonality coefficient because it quantifies how tone-like a sound is, which is the amount of peaks in a power spectrum
* spectral rolloff - frequency below which a specified percentage of the total spectral energy lies
* tonnetz - lattice diagram representing tonal space
* zero crossing rate - the rate at which the signal changes from positive to negative or reverse
* tempogram - local autocorrelation of the onset strength envelope, returns an array
* fourier tempogram - 


Most of the above features from librosa still output arrays, so different song durations will result in different array lengths. We could use the same method as above where we either trim or pad values, but then we would be creating even more data than the 2 Tbs calculated for the raw data signal.  

Instead what we could do is take primitives of these transformed arrays (i.e. the min, max, std, etc), or we can could use a large/auto-adjusting window size for the features calculations mentioned above. Let's make the transformations to a signal audio track and then we can think of primitives to take on those transformations.

```python
aud, sr = librosa.load(track2)

chroma = librosa.feature.chroma_stft(aud, sr)
cqt = librosa.feature.chroma_cqt(aud, sr)
cens = librosa.feature.chroma_cens(aud, sr)
mel = librosa.feature.melspectrogram(aud, sr)
mfcc = librosa.feature.mfcc(aud, sr)
rms = librosa.feature.rms(aud, sr)
spec_cntr = librosa.feature.spectral_centroid(aud, sr)
spec_band = librosa.feature.spectral_bandwidth(aud, sr)
spec_ctrst = librosa.feature.spectral_contrast(aud, sr)
spec_flat = librosa.feature.spectral_flatness(aud)
spec_roll = librosa.feature.spectral_rolloff(aud, sr)
tonz = librosa.feature.tonnetz(aud, sr)
zero_xr = librosa.feature.zero_crossing_rate(aud, sr)
```

```python
def calculate_primitives(feature_name, values):
    feature_dict = dict()
    feature_dict[name + 'mean'] = np.mean(values, axis=1)
    feature_dict[name + 'std'] = np.std(values, axis=1)
    feature_dict[name + 'skew'] = stats.skew(values, axis=1)
    feature_dict[name + 'kurtosis'] = stats.kurtosis(values, axis=1)
    feature_dict[name + 'median'] = np.median(values, axis=1)
    feature_dict[name + 'min'] = np.min(values, axis=1)
    feature_dict[name + 'max'] = np.max(values, axis=1)
    return feature_dict
```
